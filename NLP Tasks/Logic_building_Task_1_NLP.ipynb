Task 1: Conceptual Understanding





Q1: Difference between 'Love' and 'love' in NLP?

In NLP, "Love" and "love" are treated as different words because most systems are case-sensitive.
"Love" (capital L) may represent a proper noun (like a name or title) "love" (small l) usually represents a verb or emotion
Because of this, preprocessing often includes converting all text to lowercase so both forms are treated the same.



Q2: What happens if stopwords are not removed?
Stopwords are common words such as "is", "the", "and", "in".
If they are not removed:
The text contains unnecessary words (noise) Models take longer to process data Important words may lose importance Accuracy can decrease in some tasks
Example: "The movie is very good" After removing stopwords: "movie very good"



Q3: When is removing stopwords harmful?
Scenario 1: Sentiment Analysis
Sentence: "This is not good" After removing stopwords: "good"
The meaning changes from negative to positive, which is incorrect.
Scenario 2: Question Answering / Search Systems
Sentence: "What is the best laptop?" After removing stopwords: "best laptop"
The question structure is lost, which can affect understanding and response accuracy.



Q4: Stemming vs Lemmatization?
Stemming: Reduces words to their root form by cutting off parts Does not always produce valid words
Examples:
"running" → "run" "studies" → "studi" (not a proper word)
Lemmatization: Converts words to their correct base or dictionary form Uses grammar and vocabulary
Examples:
"running" → "run" "studies" → "study"

Task 2: Preprocessing Function

```python
import re

def preprocess_text(text):
    """
    Advanced NLP preprocessing function

    Steps:
    - Lowercasing
    - Remove URLs & emails
    - Remove numbers
    - Handle repeated characters
    - Remove special characters & emojis
    - Remove extra spaces
    - Tokenization
    - Remove short tokens (≤2) except 'no', 'not'
    """

    # 1. Handle edge cases
    if not text or not isinstance(text, str):
        return [], ""

    # 2. Convert to lowercase
    text = text.lower()

    # 3. Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # 4. Remove email-like patterns
    text = re.sub(r"\S+@\S+", "", text)

    # 5. Remove numbers
    text = re.sub(r"\d+", "", text)

    # 6. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 7. Remove special characters (including emojis)
    text = re.sub(r"[^\w\s]", "", text)

    # 8. Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # 9. Tokenization
    tokens = text.split()

    # 10. Remove short tokens (≤2) except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    # 11. Create cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence


test_cases = [
    "I have 2 dogs",                      # numbers
    "This   is   good",                   # extra spaces
    "soooo goooood!!!",                  # repeated chars
    "WOW!!! This is GREAT!!!",           # lowercase
    "Visit http://example.com now",      # URL
    "Contact me at test@gmail.com",      # email
    "No I am not happy",                 # keep 'no', 'not'
    "OK hi go",                          # remove short words
]

for text in test_cases:
    tokens, cleaned = preprocess_text(text)
    print("Original:", text)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-" * 40)
```

Task 4: Token Analytics

```python
import numpy as np

print("🔍 Token Analytics Results\n")

for i, text in enumerate(sample_inputs, 1):
    tokens, cleaned = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = np.mean([len(t) for t in tokens]) if tokens else 0

    print(f"Sentence {i}")
    print(f"Original Text        : {text}")
    print(f"Tokens               : {tokens}")
    print(f"Total Tokens         : {total_tokens}")
    print(f"Unique Tokens        : {unique_tokens}")
    print(f"Average Token Length : {round(avg_token_length, 2)}")
    print("-" * 60)
```

Task 5: Frequency Analysis

```python
from collections import Counter

# Step 1: Collect all tokens
all_tokens = []

for text in sample_inputs:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

# Step 2: Count frequencies
word_counts = Counter(all_tokens)

# Step 3: Top 10 most frequent words
print("🔝 Top 10 Most Frequent Words:\n")
for word, count in word_counts.most_common(10):
    print(f"{word} : {count}")

# Step 4: Top 5 least frequent words
print("\n🔻 Top 5 Least Frequent Words:\n")

# Sort ascending by frequency
least_common = sorted(word_counts.items(), key=lambda x: x[1])

for word, count in least_common[:5]:
    print(f"{word} : {count}")
```

Task 6: Full Pipeline

```python
# Task 6: Full Pipeline

def full_pipeline(text_list):
    """
    Full NLP preprocessing pipeline

    Input:
        text_list -> list of raw text sentences

    Output:
        Dictionary with:
        - all tokens combined
        - cleaned sentences
    """

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Call the function OUTSIDE
output = full_pipeline(sample_inputs)

# Print results
print("All Tokens:\n", output["tokens"])
print("\nCleaned Sentences:\n", output["clean_sentences"])
```
