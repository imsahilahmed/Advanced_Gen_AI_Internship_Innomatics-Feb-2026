Problem 1: Mobile Recharge Validation System

 Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps


Payment gateways


Hint
Use function for validation


Use while loop for retry


```python
def validate():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))


        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again!\n")

        elif amount not in valid_plans:
            print("Invalid plan selected. Available plans are:", valid_plans, "\n")

        else:
            print("Recharge successful!")
            break

validate()

```

Problem 2: Inventory Reorder Alert System

Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use
Warehouse systems


Retail management


Hint
Use function


Use for loop on dictionary







```python
def inventory(products):
  for product,stock in products.items():
    if stock<15:
      print(f"{product} --> Reorder again")
    else:
      print(f"{product} --> Stock OK")
products={'Mobile':12,'Laptop':40,'Gpu':70,'Mac':14,'Ipad':50}
inventory(products)
```

 Problem 3: Student Result Processing System

Problem Statement
Create a function that:
Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint
Use loop to calculate total


Return result from function


```python
def result_processing(marks):
  total=0
  for mark in marks:
    total+=mark
  avg_marks=total/len(marks)
  if avg_marks>=50:
    print("Pass")
  else:
    print("Fail")

marks=[50,60,40,70,60,70]
result_processing(marks)
```

Problem 4: Cab Fare Estimator with Retry Option

Problem Statement
Create a cab fare calculator:
Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


Real-Time Use
Ride-hailing apps


Transport systems


Hint
Function for fare calculation


While loop for retry


```python
def calculate_fare(distance, is_peak_hour):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if is_peak_hour:
        fare = fare + (fare * 0.25)

    return fare

def cab_fare_estimator():
    print("=" * 40)
    print("       CAB FARE ESTIMATOR ")
    print("=" * 40)

    while True:
        # Get distance input
        distance = float(input("\nEnter distance in km: "))

        # Get peak hour input
        peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
        is_peak_hour = peak_input == "yes"

        # Calculate and display fare
        fare = calculate_fare(distance, is_peak_hour)

        print("\n--- Fare Breakdown ---")
        print(f"  Base Fare          : ₹50")
        print(f"  Distance Charge    : ₹{distance} km × ₹12 = ₹{distance * 12:.2f}")

        if is_peak_hour:
            basic = 50 + (distance * 12)
            print(f"  Peak Hour Surcharge: ₹{basic * 0.25:.2f} (25% extra)")

        print(f"  {'Peak Hour' if is_peak_hour else 'Normal Hour'} Applied")
        print(f"  -------------------------")
        print(f"  Total Fare         : ₹{fare:.2f}")

        # Retry option
        print("\n" + "=" * 40)
        retry = input("Do you want to estimate another fare? (yes/no): ").strip().lower()

        if retry != "yes":
            print("\nThank you for using Cab Fare Estimator!")
            print("=" * 40)
            break

cab_fare_estimator()
```

Problem 5: Employee Attendance Eligibility Checker

Problem Statement
Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


Real-Time Use
HR payroll systems


Internship tracking


 Hint
Use loop to count present days


Use return statement





```python
# Employee Attendance Eligibility Checker

def check_eligibility(name, attendance):
    total_days = len(attendance)
    present_days = 0

    # Count present days using loop
    for status in attendance:
        if status == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        result = "Eligible"
    else:
        result = " Not Eligible"

    # Display Report
    print("-" * 40)
    print(f"  Employee Name   : {name}")
    print(f"  Total Days      : {total_days}")
    print(f"  Present Days    : {present_days}")
    print(f"  Absent Days     : {total_days - present_days}")
    print(f"  Attendance %    : {percentage:.2f}%")
    print(f"  Status          : {result}")
    print("-" * 40)

    return result




# Employee Data
employees = {
    "Sahil":    ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"],
    "Muneer":   ["A", "A", "P", "A", "P", "A", "A", "P", "A", "A"],
    "Akheel":   ["P", "P", "P", "P", "P", "P", "A", "A", "P", "P"],
    "Akram":    ["A", "P", "A", "P", "A", "P", "A", "P", "A", "P"],
    "Amer":     ["P", "P", "P", "A", "P", "P", "P", "P", "A", "P"],
}

for emp_name, att_list in employees.items():
    check_eligibility(emp_name, att_list)
```

Problem 6: Password Strength Checker

Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use
Login systems


Security validation


Hint
Use loop over string


Use conditions inside function



```python
def check_password_strength(password):
    special_characters = "@#$"

    has_min_length = False
    has_digit = False
    has_special = False

    # Check minimum length
    if len(password) >= 8:
        has_min_length = True

    # Loop over each character to check digit and special char
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True

    # Display individual rule results
    print(f"\n  Password         : {password}")
    print(f"  Length Check     : {'Passed' if has_min_length else 'Failed'} (Min 8 chars)")
    print(f"  Digit Check      : {'Passed' if has_digit     else 'Failed'} (At least 1 digit)")
    print(f"  Special Char     : {'Passed' if has_special   else 'Failed'} (@ # $ required)")

    # Final strength result
    if has_min_length and has_digit and has_special:
        strength = "Strong Password"
    elif has_min_length and (has_digit or has_special):
        strength = "Moderate Password"
    else:
        strength = "Weak Password"

    print(f"  Strength         : {strength}")
    print("-" * 45)

    return strength


# Main Program
print("=" * 45)
print("       PASSWORD STRENGTH CHECKER")
print("=" * 45)

# Test Passwords
passwords = [
    "sahil",
    "sahil@",
    "sahil23820",
    "sahil@123",

]

for pwd in passwords:
    check_password_strength(pwd)
```
