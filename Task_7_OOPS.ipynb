Problem 1: Smart Light Controller

Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.
üåç Real-Time Use
Smart home automation systems
üì• Sample Input
Light Name: Bedroom Light
Action: ON
üì§ Expected Output
Bedroom Light is ON


```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def control(self, action):
        if action.upper() == "ON":
            self.turn_on()
        elif action.upper() == "OFF":
            self.turn_off()
        else:
            print("Invalid action! Use 'ON' or 'OFF'.")
            return

        self.display_status()

    def display_status(self):
        print(f"{self.name} is {self.status}")
light_name = "Bedroom Light"
action = "ON"

light = SmartLight(light_name)
light.control(action)
```

Problem 2: Employee ID Card System

 Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.
Real-Time Use
Corporate HR systems
üì• Sample Input
Employee Name: Rahul
Employee ID: EMP102
Department: IT

üì§ Expected Output
Employee ID Card
Name: Rahul
ID: EMP102
Department: IT


```python
class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.details = {
            "Name"       : name,
            "ID"         : emp_id,
            "Department" : department
        }

    def display(self):
        for key, value in self.details.items():
            print(f"{key:<12}: {value}")
emp = EmployeeIDCard("Rahul", "EMP102", "IT")
emp.display()
```

Problem 3: Mobile Contact Record

Problem Statement
Create a class to store contact name and phone number and display contact information.
Real-Time Use
Mobile contact applications
üì• Sample Input
Contact Name: Anita
Phone Number: 9876543210
üì§ Expected Output
Contact Saved
Name: Anita
Phone: 9876543210


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def __str__(self):
              return (f"Name  : {self.name}\n"
                     f"Phone : {self.phone}\n")

contact = Contact("Anita", "9876543210")
print(contact)
```


Problem 4: Product Price Tag Generator
Problem Statement
Create a class that stores product name and price and prints a formatted price tag.
 Real-Time Use
Retail billing systems
üì• Sample Input
Product Name: Headphones
Price: 2499
üì§ Expected Output
Product: Headphones
Price: ‚Çπ2499


```python
class PriceTag:
    def __init__(self, product, price):
        self.product = product
        self.price = price

    def __str__(self):
        return (
                f"Product : {self.product}\n"
                f"Price   : ‚Çπ{self.price}\n")

tag = PriceTag("Headphones", 2499)
print(tag)
```

Problem 5: Movie Rating Display System

Problem Statement
Create a class that stores movie name and rating and displays movie details.
Real-Time Use
Streaming platforms
üì• Sample Input
Movie Name: Inception
Rating: 4.8
üì§ Expected Output
Movie: Inception
Rating: 4.8 / 5



```python
class Movie:
    def __init__(self, name, rating):
        self.name   = name
        self.rating = rating

    def __str__(self):
        return (
                f"Movie  : {self.name}\n"
                f"Rating : {self.rating} / 5\n")

movie = Movie("Inception", 4.8)
print(movie)
```

Problem 6: Delivery Address Manager

Problem Statement
Create a class that stores customer name and delivery address and prints delivery details.
Real-Time Use
Courier and logistics platforms
üì• Sample Input
Customer Name: Suman
Address: Hyderabad
üì§ Expected Output
Delivery Details
Customer: Suman
Address: Hyderabad




```python
class DeliveryAddress:
    def __init__(self, customer, address):
        self.customer = customer
        self.address  = address

    def __str__(self):
        return (
                f"Customer : {self.customer}\n"
                f"Address  : {self.address}\n"
                )


delivery = DeliveryAddress("Suman", "Hyderabad")
print(delivery)
```
