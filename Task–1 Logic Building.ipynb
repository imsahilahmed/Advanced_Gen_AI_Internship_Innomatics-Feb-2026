1.User Login Check

Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"

Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"


Real-World Application: Authentication systems

2. Pass / Fail Analyzer

Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks â‰¥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly

Real-World Application: Academic evaluation systems


Here i used loop to iterate the marks of students and check the constraint of passing using if-else conditions

```python
marks=[45, 78, 90, 33, 60]
pass_students=0
fail_students=0

for i in marks:
  if i>=50:
    pass_students+=1
  else:
    fail_students+=1
print("Academic evaluation systems")

print("Number of pass students:", pass_students)

print("Number of fail students:", fail_students)

```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list

Real-World Application: Data preprocessing before analysis


In below cleaning of list , i used simple list comprehension and used replace to remove spaces, and lower() to convert string to lowercase

```python
names=[" Alice ", "bob", " CHARLIE "]
cleaned_names=[name.replace(" ", "") for name in names]

cleaned_list=[name.lower() for name in cleaned_names]

print("Data preprocessing before analysis")
print("The Preprocessed List after cleaning: ",cleaned_list)
```


4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters

Real-World Application: Text filtering and validation systems


Here i used simple for loop and if else statements

```python
messages=["Hi", "Welcome to the platform", "OK"]

print("Text filtering and validation systems")

for message in messages:
  if len(message)>10:
    print(f'The message "{message}" is longer')
  else:
    print(f'The message "{message}" is not long')
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count

Real-World Application: Monitoring and log analysis systems


Down below i use python built in function count() to count number of provided elements in the list

```python
logs=["INFO", "ERROR", "WARNING", "ERROR"]

Total_error_counts=logs.count("ERROR")
print(" Monitoring and log analysis systems")
print("The total number of error entries are:",Total_error_counts)
```
